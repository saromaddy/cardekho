# Merging all the excel files into single file

```python
import pandas as pd
import os

# Folder containing your Excel files
folder_path = "csv files"

# Get all Excel files from the folder
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Create an empty list to store the dataframes
df_list = []

# Loop through each file and read them into a DataFrame
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    df_list.append(df)

# Concatenate all dataframes
combined_df = pd.concat(df_list, ignore_index=True)

# Write the combined dataframe to a single Excel file
output_path = "merged/combined_excel_file.xlsx"
combined_df.to_excel(output_path, index=False)

print(f"All files combined into {output_path}")
```

```python
columns_with_dict = ['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs']
```

```python
# Function to expand dictionary columns
def expand_dict_columns(df, column):
    df[column] = df[column].apply(lambda x: ast.literal_eval(str(x)) if isinstance(x, str) else x)
    expanded_columns = pd.json_normalize(df[column])
    expanded_columns.columns = [f"{column}_{col}" for col in expanded_columns.columns]
    return expanded_columns

```

```python
# Iterate over the dictionary columns and expand them
for col in columns_with_dict:
    expanded_df = expand_dict_columns(df, col)
    df = df.drop(columns=[col])  # Drop original column
    df = pd.concat([df, expanded_df], axis=1)  # Concatenate expanded columns

# Save the updated dataframe to a new Excel file
df.to_excel("expanded_columns_excel_file.xlsx", index=False)

print("Columns expanded and saved to 'expanded_columns_excel_file.xlsx'")
```

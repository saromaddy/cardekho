# Merging all the excel files into single file

```python
import pandas as pd
import json
```

```python
df_bangalore = pd.read_excel("csv files/bangalore_cars.xlsx")
df_bangalore["City"] = 'Bangalore'
df_chennai = pd.read_excel("csv files/chennai_cars.xlsx")
df_chennai["City"] = 'Chennai'
df_delhi = pd.read_excel("csv files/delhi_cars.xlsx")
df_delhi["City"] = "Delhi"
df_hyderabad = pd.read_excel("csv files/hyderabad_cars.xlsx")
df_hyderabad["City"] = "Hyderabad"
df_jaipur = pd.read_excel("csv files/jaipur_cars.xlsx")
df_jaipur["City"] = "Jaipur"
df_kolkata = pd.read_excel("csv files/kolkata_cars.xlsx")
df_kolkata["City"] = "Kolkata"

merged_df = pd.concat([df_bangalore,df_chennai,df_delhi,df_hyderabad,df_jaipur,df_kolkata], ignore_index=True)
```

```python
data = pd.read_json("merged/merged_data.json")
```

```python
df_newcar = data["new_car_detail"]
df_newcar
```

```python
df_new_car_detail = pd.DataFrame(df_newcar)
df_new_car_detail
```

```python
import ast
df_new_car_detail['new_car_detail'] = df_new_car_detail['new_car_detail'].apply(ast.literal_eval)

# Normalize the 'new_car_detail' data
car_details_normalized_data = pd.json_normalize(df_new_car_detail['new_car_detail'])

car_details_normalized_data
```

```python
data.columns
```

```python
df_overview = data["new_car_overview"]
df_new_car_overview = pd.DataFrame(df_overview)

df_new_car_overview['new_car_overview'] = df_new_car_overview['new_car_overview'].apply(ast.literal_eval)

# Extract the 'top' data and create columns from the keys
top_data = df_new_car_overview['new_car_overview'].apply(lambda x: {item['key']: item['value'] for item in x['top']})

# Convert the dictionary of 'top' data to a DataFrame
car_overview_normalized_data = pd.DataFrame(top_data.tolist())

# Display the resulting DataFrame
car_overview_normalized_data
```

```python
df_car_feature = data['new_car_feature']
df_new_car_feature = pd.DataFrame(df_car_feature)

df_new_car_feature
```

```python

# Ensure 'new_car_feature' is a dictionary (convert from string if needed)
if df_new_car_feature['new_car_feature'].apply(type).eq(str).any():
    df_new_car_feature['new_car_feature'] = df_new_car_feature['new_car_feature'].apply(ast.literal_eval)

# Initialize an empty dictionary to hold column data
columns = {}

# Process the 'top' data
df_new_car_feature['top'] = df_new_car_feature['new_car_feature'].apply(lambda x: x.get('top', []))
columns['Features'] = df_new_car_feature['top'].apply(len)

# Process the 'data' field
df_new_car_feature['data'] = df_new_car_feature['new_car_feature'].apply(lambda x: x.get('data', []))
for entry in df_new_car_feature['data'].values[0]:
    heading = entry['heading']
    list_length = len(entry['list'])
    columns[f'{heading}'] = [list_length] * len(df_new_car_feature)

# Convert the dictionary to a DataFrame
car_feature_normalized_data = pd.DataFrame(columns)

# Display the DataFrame
car_feature_normalized_data
```

```python
df_newspecs = data['new_car_specs']
df_new_car_specs = pd.DataFrame(df_newspecs)

# Ensure 'new_car_specs' column is a dictionary
df_new_car_specs['new_car_specs'] = df_new_car_specs['new_car_specs'].apply(ast.literal_eval)

# Normalize the 'new_car_specs' data directly
normalized_data = df_new_car_specs['new_car_specs'].apply(
    lambda specs: {
        **{item['key']: item['value'] for item in specs['top']},
        **{item['key']: item['value'] for section in specs['data'] for item in section['list']}
    }
)

# Convert to DataFrame
car_specification1_normalized_data = pd.json_normalize(normalized_data)

car_specification1_normalized_data
```

```python
city = data['City']
df_city = pd.DataFrame(city)
df_city
```

```python
full_expanded = pd.concat([car_details_normalized_data,car_overview_normalized_data,car_feature_normalized_data,car_specification1_normalized_data,df_city],axis = 1)
full_expanded
```

```python
def steer(steer_type):
    if pd.isna(steer_type):
        return steer_type
    steer_type_lower = steer_type.lower()
    
    if 'electronic' in steer_type_lower or 'electric' in steer_type_lower or 'electrical' in steer_type_lower:
        return 'Electronic'
    elif 'epas' in steer_type_lower:
        return 'EPAS'
    elif 'power' in steer_type_lower:
        return 'Power'
    elif 'manual' in steer_type_lower:
        return 'Manual'
    else:
        return 'Other'

df['Steering Type'] = df['Steering Type'].apply(steer)

df['Steering Type'].value_counts()
df["Steering Type"].fillna(df["Steering Type"].mode()[0],inplace=True)
# full_expanded['Steering Type'] = full_expanded['Steering Type'].fillna('Power')
df['Steering Type'].isnull().sum()

```

```python
def convert_amount(amount):
    # Remove '₹' and whitespace
    amount = amount.replace('₹ ', '')
    
    # Handle Crore
    if 'Crore' in amount:
        value = float(re.sub(r'[^\d.]', '', amount))
        return int(value * 10000000)  # Convert crore to integer
    
    # Handle Lakh
    elif 'Lakh' in amount:
        value = float(re.sub(r'[^\d.]', '', amount))
        return int(value * 100000)  # Convert lakh to integer
    
    # Handle plain amount
    else:
        value = int(re.sub(r'[^\d]', '', amount))
        return value

df['price'] = df['price'].apply(convert_amount)
df['price']
```

```python
def categorize_brake_type(brake_type):
    if pd.isna(brake_type):
        return brake_type
    brake_type_lower = brake_type.lower()
    
    if 'disc' in brake_type_lower or 'disk' in brake_type_lower:
        return 'Disc'
    elif 'drum' in brake_type_lower:
        return 'Drum'
    elif 'hydraulic' in brake_type_lower or 'mechanical' in brake_type_lower:
        return 'Hydraulics'
    elif 'electric' in brake_type_lower or 'parking' in brake_type_lower:
        return 'Electric'
    else:
        return 'Other'

df['Rear Brake Type'] = df['Rear Brake Type'].apply(categorize_brake_type)
df['Rear Brake Type'].value_counts()
```

```python
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]
    return df

# Apply outlier removal
df = remove_outliers(df, cont)

print(df.shape)
```
